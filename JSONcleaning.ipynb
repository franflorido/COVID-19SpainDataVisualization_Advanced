# Importing Json file and cleaning data

In this notebook we are going to clean the COVID json file I obtained from:
    https://covid19.secuoyas.io/api/v1/es/ccaa?fecha=2020-04-01

```python
import pandas as pd
import numpy as np
import json
```

```python
data = json.load(open('D:/Documentos/Master/temariomaster/p10.Visualizacion/Tareas/Tarea1/src/ccaa.json'))
data
```

```python
ccaa = data['timeline'][0]['regiones']
ccaa
```

```python
for element in ccaa:
    element.pop('nivelAdministrativo', None)
    element.pop('codigoIsoLugar', None)
    element.pop('long', None)
    element.pop('lat', None)
    element['data'].pop('_id',None)
    element['data'].pop('casosUci',None)
    element['data'].pop('casosFallecidos',None)
    element['data'].pop('casosHospitalizados',None)
    element['data'].pop('casosRecuperados',None)
    element['data'].pop('casosConfirmadosDiario',None)
    element['data'].pop('casosUciDiario',None)
    element['data'].pop('casosFallecidosDiario',None)
    element['data'].pop('casosHospitalizadosDiario',None)
    element['data'].pop('casosRecuperadosDiario',None)
```

```python
ccaa
```

```python
for element in ccaa:
    element['data']=element['data']['casosConfirmados']
```

```python
def writeToJSONFile(path, fileName, data):
    filePathNameWExt =  path + '/' + fileName + '.json'
    with open(filePathNameWExt, 'w') as fp:
        json.dump(data, fp)
```
